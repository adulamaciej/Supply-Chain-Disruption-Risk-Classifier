EXPLORATORY DATA ANALYSIS

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/suppliers.csv")
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

```python
df["contract_type"]  = df["contract_type"].astype("category")
df["supplier_tier"]  = pd.Categorical(df["supplier_tier"], categories=["Tier 1", "Tier 2", "Tier 3"], ordered=True)
df["transport_mode"] = df["transport_mode"].astype("category")

#CSV doesn't preserve categorical dtype — it saves everything as strings, 
# so to re-casting after loading is needed, that's why it is not in the first notebook
```

```python
labels = {0: "None", 1: "Low", 2: "Medium", 3: "High", 4: "Critical"}
df["risk_label"] = df["risk_level"].map(labels)

counts = df["risk_level"].value_counts().sort_index()
for k, v in counts.items():
    print(k, labels[k], v)
```

1200 suppliers across 5 risk levels, with Critical (384) most common and Low (131) least common. Class imbalance will be addressed with class_weight="balanced" in all models during training.

```python
medium_archetype = df[df["archetype"] == 2]
print(medium_archetype["risk_level"].value_counts().sort_index())
```

Out of 400 Medium archetype suppliers, 210 were correctly classified as Medium. 61 leaked to Low, 66 to High, 63 to Critical.
